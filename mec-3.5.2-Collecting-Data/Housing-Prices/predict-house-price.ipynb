## Clean Data. Handle Nulls.

### does null data represent information, or lack of information?

```python
# create list of null columns
null_cols = []
for col in df.columns:
    if df[col].isnull().values.any() == True:
        null_cols.append(col)
null_cols    
```

```python
# 'LotFrontage'
# because there are no 0's, 
# I am assuming NaN means 0 linear feet of house lining street
df[df.LotFrontage == 0].size
```

```python
# 'LotFrontage' fill null with 0
print(df.LotFrontage.isna().sum())
df.LotFrontage.fillna(0, inplace=True) #fill with null values
print(df.LotFrontage.isna().sum())
```

```python
# Replace nulls in 'Alley' with 'NoAccess'
df.Alley.fillna('NoAccess', inplace=True)
```

```python
# Replace 'MasVnrType' nulls with 'NoMasonry'
df.MasVnrType.fillna('NoMasonry', inplace=True)
```

```python
# Replace 'MasVnrArea' nulls with 0
df.MasVnrArea.fillna(0.0, inplace=True)
```

```python
# Replace Basement nulls with 'NB' for No Basement
df.BsmtQual.fillna('NB', inplace=True)
```

```python
df.BsmtCond.fillna('NB', inplace=True)
df.BsmtExposure.fillna('NB', inplace=True)
df.BsmtFinType1.fillna('NB', inplace=True)
df.BsmtFinType2.fillna('NB', inplace=True)
```

```python
# Electrical, impute the 1 null with the most common electrical frameworke
most_common = df.Electrical.value_counts().index[0]
df.Electrical.fillna(most_common, inplace=True)
```

```python
# FireplaceQu replace null with 'NF' for No Fireplace
df.FireplaceQu.fillna('NF', inplace=True)
```

```python
# NG for No Garage
df.GarageType.fillna('NG', inplace=True)
df.GarageYrBlt.fillna('NG', inplace=True)
df.GarageFinish.fillna('NG', inplace=True)
df.GarageQual.fillna('NG', inplace=True)
df.GarageCond.fillna('NG', inplace=True)
```

```python
# NP for No Pool
df.PoolQC.fillna('NP', inplace=True)
```

```python
# NF for No Fence
df.Fence.fillna('NF', inplace=True)
```

```python
# NMF for No MiscFeature
df.MiscFeature.fillna('NMF', inplace=True)
```
