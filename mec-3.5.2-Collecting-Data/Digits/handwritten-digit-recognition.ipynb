## Handwritten Digit Recognition
https://www.kaggle.com/competitions/digit-recognizer/data

gather handwritten digit data and display images

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
digits = pd.read_csv('train.csv')
digits.info()
```

```python
# are any values missing from the DF?
digits.isnull().values.any()
```

```python
labels = digits['label']
labels
```

```python
digits = digits.drop('label', axis=1)
digits
```

```python
# Reshape to 28 * 28 so we can see the images
images = digits.values.reshape(-1, 28, 28, 1)
images.shape
```

```python
# plt image
plt.imshow(images[0])
```

```python
num = 25
num_row = 5
num_col = 5

# plot images
fig, axes = plt.subplots(num_row, num_col, figsize=(15,15))
for i in range(num):
    ax = axes[i//num_col, i%num_col]
    ax.imshow(images[i])
    ax.set_title('Label: {}'.format(labels[i]), color='white', fontsize=14)
    ax.tick_params(axis='both', which='both', left=False, bottom=False, labelleft=False, labelbottom=False)
```
